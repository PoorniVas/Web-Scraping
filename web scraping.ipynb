```python
import requests
```

```python
url="https://www.flipkart.com/computers/storage/pen-drives/pr?sid=6bo%2Cjdy%2Cuar"
```

```python
permit=requests.get(url)
```

```python
permit
```

```python
text_url=BeautifulSoup(permit.text)
```

```python
Brand=text_url.select('.s1Q9rs')
```

```python
import pandas as p
pendrive_list=p.DataFrame(Brand,columns=["Brand Name"])
```

```python
Price=text_url.select("._8VNy32 ._30jeq3")
```

```python
pendrive_list['Offered Price']=Price
```

```python
rating=text_url.select('._1lRcqv ._3LWZlK')
```

```python
pendrive_list['Rating']=rating
```

```python
color=text_url.select('._3Djpdu')
```

```python
pendrive_list['Color']=color
```

```python
buyers=text_url.select('._2_R_DZ')
```

```python
pendrive_list['Buyers']=buyers
```
